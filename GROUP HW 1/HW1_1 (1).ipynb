```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = load_iris(as_frame=True)
X, y = iris.data, iris.target
print(iris.DESCR)
```

```python
# STEP 2
k_values = range(1, 21)
avg_accuracies = [cross_val_score(KNeighborsClassifier(n_neighbors=k), X, y, cv=5).mean() for k in k_values]
```

```python
# STEP 2
plt.plot(k_values, avg_accuracies)
plt.xlabel('Value of K')
plt.ylabel('Average Accuracy')
plt.title('Average Accuracy for Different Values of K')
plt.grid(True)
plt.xticks(range(1, 21, 2))
plt.show()
```

```python
df = pd.DataFrame(avg_accuracies, columns=['Average_accuracy'])
```

```python
df.index.name = "K_value"
df
```

```python
# STEP 3 (For loop line chart)
iris = load_iris(as_frame=True)
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=5)

k_values_list = list(range(1, 21))
accuracy_scores = []

for k in k_values_list:
    kNN = KNeighborsClassifier(n_neighbors=k)
    kNN.fit(X_train, y_train)
    accuracy = kNN.score(X_test, y_test)
    accuracy_scores.append(accuracy)


plt.plot(k_values_list, accuracy_scores, marker='o')
plt.title('Accuracy vs. k-Neighbors')
plt.xlabel('Value of K')
plt.ylabel('Accuracy')
plt.grid(True)
plt.xticks(k_values_list)
plt.show()

print(accuracy_scores)
```
